
# 📘 파이썬 구조와 실행 개념 요약

## ✅ 1. 버튼과 코드의 차이
- **물리적인 버튼의 한계**  
  : 제한된 개수의 동작만 가능함.
- **코드의 장점**  
  : 무한한 수의 "버튼(명령)"을 만들 수 있음.

---

## ✅ 2. 파일과 코드의 관계
- **파일**은 순서대로 실행되는 코드의 집합.
- 코드를 모아두면 **자동화**가 가능해짐.

---

## ✅ 3. 프로그래밍의 흐름
- **프로그래머** → **프로그래밍 언어** → **프로그래밍** → **프로그램**

> 💡 프로그램이란?  
> 컴퓨터가 가지고 있는 기능을 **시간의 순서에 따라 일어나도록 만든 것**.

---

## ✅ 4. 파이썬 언어의 두 가지 실행 방식
- **셸(Shell)**  
  - 한 줄 입력 및 실험적/간단한 작업용
- **파일(File)**  
  - 자주 실행되는 작업을 저장해두고 필요할 때 실행

---

## ✅ 5. 파이썬 실행 구조

```
셸 또는 파일 입력
        ↓
  파이썬 인터프리터
        ↓
  운영체제(System)
        ↓
   컴퓨터 실행
```

- **인터프리터**란?  
  코드를 읽고 해석해서 **운영체제에 전달**하는 번역기 역할
