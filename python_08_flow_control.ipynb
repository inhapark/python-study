# 📘 파이썬 제어문과 비교 연산자

🐍 **1. 제어문 (Flow Control)**
- 조건문 (`if`, `else`, `elif`)  
- 반복문 (`for`, `while`)  
✅ 프로그램의 흐름을 조건에 따라 제어

---

🐍 **2. 비교 연산자 (Comparison Operator)**
- 데이터를 비교하여 `True` 또는 `False` 값을 반환
- 조건문의 핵심 역할

📌 예제:
```python
print(True)    # 출력: True
print(False)   # 출력: False
```

📌 비교 연산자 예제:
```python
print(1 > 2)     # False
print(1 == 1)    # True
```

✅ 비교 연산자는 두 개의 피연산자를 비교함  
- `==` : 같다  
- `>=` : 이상 (크거나 같다)  

---

👉 핵심 정리표

| 연산자 | 의미 |
|--------|------|
| `==`   | 같다 |
| `!=`   | 다르다 |
| `>`    | 크다 |
| `<`    | 작다 |
| `>=`   | 크거나 같다 |
| `<=`   | 작거나 같다 |
