# 📘 디버깅(Debugging)의 개념과 방법

🐍 **1. 버그(Bug)**  
- 프로그램 내 **예상치 못한 오류**를 의미함  
✅ 정상적인 흐름을 방해하며, 반드시 찾아서 해결해야 함

---

🐍 **2. 디버깅(Debug)**  
- 버그를 **찾고 제거하는 과정**  
✅ 소프트웨어 개발 과정에서 매우 중요한 단계

---

🐍 **3. 디버거(Debugger)**  
- 버그를 찾고 수정할 수 있도록 돕는 **강력한 도구**  
✅ 코드의 흐름을 추적하고 중단점 설정, 변수 값 확인 가능

---

🐍 **4. 버그를 찾는 방법 2가지**  
1. **반씩 나눠서 직접 실행**  
   - 데이터를 절반씩 줄여가며 실행하여 **문제가 발생한 지점**을 좁혀가는 방식  
   - 예: 이진탐색처럼 문제의 원인을 추적  
2. **디버거 사용**  
   - 디버거 도구를 통해 변수 상태, 흐름 등을 **정밀하게 추적**

---

👉 핵심 정리표

| 항목 | 설명 |
|------|------|
| 버그 | 예상치 못한 프로그램 오류 |
| 디버깅 | 버그를 제거하는 과정 |
| 디버거 | 버그를 찾고 수정하는 도구 |
| 버그 찾기 방법 | 1) 반씩 나누며 실행<br>2) 디버거 사용 |
