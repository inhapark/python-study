# 📘 조건문의 중첩 (Nested if)

🐍 **1. 중첩 조건문이란?**  
- `if` 문 내부에 또 다른 `if` 문을 사용하는 구조  
✅ 여러 조건을 **순차적으로 확인**할 수 있음  
✅ 논리 구조를 **단계별로 분리**해서 처리할 때 사용

---

🐍 **2. 실행 예제**
```python
input_id = input('id:')
id = 'egoing'

input_password = input('password:')
password = '111111'

if input_id == id:
    if input_password == password:
        print('Welcome')
    else:
        print('Wrong password')
else:
    print('Wrong id')
```

📌 실행 흐름 설명:
1. 사용자에게 ID 입력 요청  
2. 입력값이 `'egoing'`이면 다음 단계로 넘어감  
3. 사용자에게 비밀번호 입력 요청  
4. 입력값이 `'111111'`이면 `"Welcome"` 출력  
5. 비밀번호가 틀리면 `"Wrong password"` 출력  
6. ID부터 틀렸으면 `"Wrong id"` 출력

---

👉 핵심 정리표

| 구문 | 설명 |
|------|------|
| `if 조건:` | 첫 번째 조건 검사 |
| `if 조건:` (내부) | 첫 조건이 맞을 때만 다음 조건 검사 |
| 중첩 조건문 | 조건 안에 조건이 들어간 구조 |
| 사용 예시 | 로그인 시 ID와 비밀번호 순차 확인 |
