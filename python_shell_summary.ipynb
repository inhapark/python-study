
# 🐍 파이썬 셸(Python Shell) 기초 요약

## ✅ 파이썬 셸의 특징
- `print()` 없이도 값이 출력됨
- `len("hello world")` → 문자열 길이 출력

```python
print(1 + 1)
1 + 1
len("hello world")
```

## ✅ 변수 여러 개 할당하기
- 세미콜론(;) 사용 → 여러 명령어를 한 줄에 실행  
- 마지막 명령어의 결과만 출력됨

```python
a = 1; b = 1; a + b
```

- 역슬래시(\) 사용 → 여러 줄에 나눠 실행할 수 있음

```python
a = 1; \
b = 1; \
a + b
```

## ✅ 셸 종료 방법
- `exit()` 또는 `Ctrl + Z` 후 `Enter`  
```python
exit()
```
