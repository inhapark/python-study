# 📘 조건문 if - elif - else

🐍 **1. 조건문의 구조**
```python
if 조건1:
    실행문1
elif 조건2:
    실행문2
else:
    실행문3
```
✅ 조건에 따라 **하나의 블록만 실행됨**

---

🐍 **2. 실행 예제: 조건별 흐름**
### 예제 1
```python
print(0)
if True:
    print(1)
elif True:
    print(2)
else:
    print(3)
print(4)
```
📌 결과:
```
0  
1  
4
```
✅ 첫 번째 조건이 True → elif, else는 무시됨

---

### 예제 2
```python
print(0)
if False:
    print(1)
elif True:
    print(2)
else:
    print(3)
print(4)
```
📌 결과:
```
0  
2  
4
```
✅ 첫 번째 조건 False → elif가 True이므로 실행

---

### 예제 3
```python
print(0)
if False:
    print(1)
elif False:
    print(2)
else:
    print(3)
print(4)
```
📌 결과:
```
0  
3  
4
```
✅ 모든 조건이 False → else 실행

---

🐍 **3. 사용자 입력을 통한 조건 분기**
```python
input_id = input('id : ') 
id1 = 'egoing'
id2 = 'basta'

if input_id == id1:
    print('Welcome')
elif input_id == id2:
    print('Welcome')
else:
    print('Who?')
```

📌 설명:  
- 사용자가 입력한 id가 `'egoing'` 또는 `'basta'`면 `"Welcome"` 출력  
- 둘 다 아니면 `"Who?"` 출력  

---

👉 핵심 정리표

| 구문 | 설명 |
|------|------|
| `if` | 첫 조건 |
| `elif` | 첫 조건이 False일 때 다음 조건 |
| `else` | 모든 조건이 False일 때 |
| `if input_id == id1` | 변수 비교 |
| `if input_id == 'egoing'` | 문자열 직접 비교 |
