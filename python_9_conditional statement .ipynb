# 📘 조건문 `if`와 실행 흐름

🐍 **1. if 조건문 기본 구조**  
```python
if 조건식:
    실행문
```
- 조건이 `True`일 때만 실행문이 실행됨  
✅ 조건이 `False`면 실행문은 무시됨

---

🐍 **2. 실행 예제 1: if 조건이 True인 경우**
```python
print(0)
if True:
    print(1)
print(2)
```
📌 결과 출력:
```
0  
1  
2
```
✅ `if True:` → 조건이 참이므로 `print(1)` 실행됨

---

🐍 **3. 실행 예제 2: if 조건이 False인 경우**
```python
print(0)
if False:
    print(1)
print(2)
```
📌 결과 출력:
```
0  
2
```
✅ `if False:` → 조건이 거짓이므로 `print(1)` 실행되지 않음

---

🐍 **4. 응용 예제: 사용자 입력을 통한 조건 비교**
```python
input_id = input('id : ')
id = 'egoing'
if input_id == id:
    print('Welcome')
```
📌 실행 흐름:
1. 사용자에게 `id : ` 라고 입력 요청  
2. 사용자가 입력한 값이 `'egoing'`이면  
3. `"Welcome"` 이 출력됨

✅ 조건에 맞는 입력이 들어와야 `if` 문이 실행됨

---

👉 핵심 정리표

| 코드 | 설명 |
|------|------|
| `if True:` | 조건이 참 → 내부 코드 실행 |
| `if False:` | 조건이 거짓 → 내부 코드 실행 안 함 |
| `input()` | 사용자 입력 받기 |
| `if input_id == id:` | 입력값이 일치하는지 비교 |

| 예시 코드 | 설명 |
|-----------|------|
| `if input_id == id:` | ✅ 변수 비교 |
| `if input_id == 'egoing':` | ✅ 문자열 직접 비교 |
