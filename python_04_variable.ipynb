# 🐍 파이썬 3-4: 변수 (Variable)

## 📘 변수란?
- 변수는 **데이터에 이름을 붙이는 것**이다.
- 같은 값을 여러 곳에서 반복해서 사용하거나, 나중에 쉽게 수정하기 위해 사용한다.


## ✅ 변수의 필요성 예시
```python
# 예: 동일한 이름이 반복되는 문장을 출력하고 싶을 때
name = 'sori'
print('hi, ' + name + ' ~ bye, ' + name + '.')
```
- 출력 결과:
```
hi, sori ~ bye, sori.
```
- 설명:
  - `name`이라는 변수에 `'sori'`를 저장한 뒤,
  - 변수 `name`을 문자열과 이어 붙여 사용하면,
  - 같은 값이 여러 번 등장할 때 **중복 없이, 실수 없이 관리**할 수 있다.


## ⭐ 변수와 문자열 결합
<span style="font-size: 150%">

```python
print('hi, ' + name + ' ~ bye, ' + name + '.')
```
- 출력 결과:
```
hi, sori ~ bye, sori.
```
- 설명:
  - 문자열은 반드시 `' '` 또는 `" "` **따옴표 안에** 있어야 하며,
  - 변수는 **따옴표 없이** 사용해야 한다.
</span>


## ✅ 변수의 활용 예: 메시지 변수 추가
```python
message = 'hi, ' + name + ' ~ bye, ' + name + '.'
print(message)
```
- 출력 결과:
```
hi, sori ~ bye, sori.
```
- 설명:
  - 중간 결과를 다시 변수 `message`에 저장하면,
  - 출력이나 반복 사용이 **더 간단해진다.**
