# 🐍 파이썬 3-3: 리스트(List) 데이터 타입

## 📘 리스트란?
- 리스트는 **서로 연관된 데이터를 모아 정리하는 도구**이다.
- 대괄호 `[]` 안에 여러 값을 넣어 하나의 변수로 묶는다.

```python
students = ['A', 'B', 'C']
```
- `students`는 리스트 변수이며, `'A'`, `'B'`, `'C'`는 **각각의 원소(element)**이다.
- 인덱스는 **0부터 시작**하며 다음과 같다:
  - `students[0]` → `'A'`
  - `students[1]` → `'B'`
  - `students[2]` → `'C'`


## ✅ 리스트에서 특정 원소 출력하기
```python
print(students[1])
```
- 출력 결과:
```
B
```
- 인덱스를 이용해 리스트 내 특정 값을 꺼낼 수 있다.


## 📘 리스트 관련 함수들

### 1️⃣ 리스트 길이 확인: `len()`
```python
print(len(students))
```
- 리스트에 몇 개의 원소가 있는지 반환한다.

---
### 2️⃣ 최솟값: `min()`, 최댓값: `max()`
```python
numbers = [3, 7, 1, 9]
print(min(numbers))  # 1
print(max(numbers))  # 9
```

---
### 3️⃣ 합계 구하기: `sum()`
```python
print(sum(numbers))  # 20
```


## 📘 리스트와 통계 모듈

### 🛠️ 모듈 불러오기
```python
import statistics
```

### 📊 평균 구하기
```python
print(statistics.mean(numbers))
```
- `statistics` 모듈은 **평균, 중앙값, 표준편차 등 통계 관련 함수**를 제공한다.
  - 평균: `statistics.mean()`
  - 중앙값: `statistics.median()`
  - 표준편차: `statistics.stdev()`


## 📘 리스트에서 랜덤값 추출

### 🛠️ 모듈 불러오기 + 사용 예시
```python
import random
print(random.choice(students))
```
- 리스트 안에서 임의의 원소 하나를 랜덤으로 추출한다.
