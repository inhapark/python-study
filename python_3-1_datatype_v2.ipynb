# 🐍 파이썬 3-1: 숫자형 데이터 타입

## 📘 숫자형 데이터 타입 개요
- 숫자형은 크게 **정수형**과 **실수형**으로 나뉜다.


## 📘 정수형 (Integer, `int`)
- 정수형은 **음수, 0, 양수**로 이루어진다.
- 표현 예시:
```python
print(-1)
print(0)
print(1)
```
- 파이썬에서 정수는 `int` 타입으로 표현된다.


## 📘 실수형 (Float, `float`)
- 실수형은 **소수점이 있는 숫자**이다.
- 표현 예시:
```python
print(1.1)
```
- 파이썬에서 실수는 `float` 타입으로 표현된다.


## ✅ 사용 예시: 문자열과 연산 출력 차이
```python
print('1+1', 1+1)  # '1+1'은 문자열 그대로 출력되고, 1+1은 계산되어 2로 출력됨
# 결과: 1+1 2
```


## 📘 연산자 (Operators)
- 더하기, 빼기, 곱하기, 나누기 등을 **연산자(operator)** 라고 한다.

| 연산 종류 | 연산자 기호 | 예시     | 결과 |
|----------|--------------|----------|------|
| 더하기   | `+`          | 1 + 2    | 3    |
| 빼기     | `-`          | 5 - 2    | 3    |
| 곱하기   | `*`          | 3 * 4    | 12   |
| 나누기   | `/`          | 10 / 2   | 5.0  |


## 📘 내장 기능: 모듈 불러오기
- 파이썬은 다양한 기능을 내장 모듈로 제공한다.
- `import`를 사용해 모듈을 불러올 수 있다.

```python
import math
print('math.sqrt(4)', math.sqrt(4))  # 제곱근 구하기
# 결과: math.sqrt(4) 2.0

print(pow(4, 2))  # 제곱 = 16
```


## 📘 랜덤 숫자 생성
```python
import random
print('random.random()', random.random())  # 0과 1 사이의 임의의 실수 출력
```
- `random.random()`은 **0과 1 사이의 소수 중 하나를 랜덤하게 반환**한다.
