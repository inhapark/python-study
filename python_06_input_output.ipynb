# 📘 파이썬 입력(Input)과 출력(Output)

🐍 **1. 입출력의 개념**  
- 프로그램의 본질은 **입력(Input) → 처리(Process) → 출력(Output)**  
- **Input 종류**: `keyboard`, `file`, `mic`, `cam`, `sensor` 등  
- **Output 종류**: `monitor`, `paper`, `file`, `speaker` 등  
✅ 입출력은 프로그램의 기본 구조를 이루는 핵심 요소임  

---

🐍 **2. 출력 (Output) 예시**  
- `print()` → 모니터(화면)에 출력  
```python
print("Hello, Python!")
```
✅ 화면에 원하는 메시지나 결과값을 출력하는 가장 기본적인 방법  

---

🐍 **3. 입력 (Input) 예시**  
- `input()` → 키보드로부터 값을 입력받음  
```python
name = input("name=")  
message = "hi, " + name + " ~ bye, " + name + "."  
print(message)
```

📌 실행 흐름  
1. `"name="` 이라는 문구가 화면에 표시됨  
2. 사용자가 직접 이름을 입력  
3. 입력한 이름이 `name` 변수에 저장됨  
4. `print(message)`가 실행되면서 입력값이 반영된 문장이 출력됨  

✅ **장점**: 데이터가 바뀌어도 코드 수정 없이 입력만 바꿔서 실행 가능  

---

👉 정리 표

| 구분 | 설명 | 예시 |
|------|------|------|
| Input | 사용자/장치가 프로그램에 주는 값 | `input()`, 키보드, 센서 |
| Process | 입력된 데이터를 처리 | 변수 할당, 계산 |
| Output | 처리 결과를 사용자/장치에 전달 | `print()`, 모니터, 스피커 |
