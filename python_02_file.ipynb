
# 📘 VS Code로 파이썬 파일 만들기 및 실행 요약노트

## 🐍 1. VS Code와 폴더 구조

- **Visual Studio Code = VS Code**  
  → 코드 편집기 프로그램
- 프로젝트를 위한 폴더를 바탕화면에 생성 (예: `python-tutorial`)
- VS Code에서 해당 폴더를 열어 파일을 관리함

✅ 폴더 구조 기반으로 코드 관리 가능

---

## 🐍 2. 파이썬 파일 생성하기

- **좌측 EXPLORER 패널**에서 **파일추가 버튼 클릭**  
  → 새 파일 생성
- 파일 이름 입력 (예: `3_2_1.py`)
- 생성된 파일을 클릭하면 코드를 작성할 수 있음

✅ 파일 안에 작성한 코드는 저장 및 재사용 가능

---

## 🐍 3. 셸과 파일 실행 방식 비교

```python
print(1 + 1)
```

- 셸에서는 입력 즉시 실행됨
- **파일은 명확하게 실행 명령을 내려야 동작함**
  - 장점: 작성한 코드를 언제든지 재사용 가능

---

## 🐍 4. 파일 실행 방법 (터미널 이용)

- 상단 메뉴: `Terminal` → `New Terminal`
- 하단에 열리는 터미널에서 다음과 같이 입력:

```bash
python 파일이름.py
```

예:
```bash
python 3_2_1.py
```

- 인터프리터가 해당 파일을 열고 내부 코드를 실행함

✅ 코드 자동 실행 및 결과 확인 가능
