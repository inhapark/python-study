# 🐍 파이썬 3-1: 데이터 타입 (Data Type)

## 📘 기본 데이터 타입
- **숫자형 (int, float, complex)**
- **문자열 (str)**
- **불(bool)**
- **None (값이 없음)**

## 📘 숫자형
- `int`: 정수 → 예: `10`, `-3`
- `float`: 실수 → 예: `3.14`, `-0.1`
- `complex`: 복소수 → 예: `2 + 3j`

## 📘 문자열
- `'hello'`, "hello" 둘 다 문자열
- 여러 줄 문자열: `''' ... '''` 또는 `""" ... """`

## 📘 불리언 (Boolean)
- `True`, `False`
- 조건문에서 많이 사용

## 📘 None
- 아무 값도 없음을 나타냄
- `null`과 유사

## ✅ 타입 확인하기
- `type(변수)` → 데이터 타입을 확인할 수 있음

```python
type(10)       # <class 'int'>
type(3.14)     # <class 'float'>
type('hi')     # <class 'str'>
type(True)     # <class 'bool'>
type(None)     # <class 'NoneType'>
```
