# 🐍 파이썬 3-2: 문자형(텍스트) 데이터 타입

## 📘 문자형 데이터 타입 개요
- 문자형 데이터 타입은 **`string` 데이터 타입(`str`)**이라고도 불린다.
- 문자열은 **작은따옴표(`'`)** 또는 **큰따옴표(`"`)**로 감싸 표현한다.


## 📘 따옴표를 이용한 문자열 출력
```python
print('hello world')
print("hello world")
```
- 출력 결과:
```
hello world
hello world
```
- 큰따옴표와 작은따옴표는 출력 결과에서 **차이가 없다**.


## 📘 여러 줄 문자열 출력
- 여러 줄로 이루어진 문자열은 삼중 따옴표(`'''` 또는 `"""`)를 사용한다.
```python
print(""" 
hello 
world
""")
```
- 출력 결과:
```
hello 
world
```


## 📘 문자열 연산

### 1️⃣ 문자열 결합
```python
print("'1'+'1'", '1'+'1')
```
- 출력 결과:
```
'1'+'1' 11
```
- `'1'+'1'`은 **문자열 결합 연산**이므로 2가 아니라 `11`이 출력된다.
- 코드 앞에 `" '1'+'1' "`처럼 큰따옴표를 붙인 이유: `'1'+'1'`을 **코드가 아니라 문자열**로 출력하기 위해서다.


### 2️⃣ 문자열 반복
```python
print('hello world'*3)
```
- 출력 결과:
```
hello worldhello worldhello world
```
- 문자열 뒤에 `*숫자`를 붙이면 해당 문자열이 **숫자만큼 반복 출력**된다.


## 📘 문자열 길이 구하기 (`len` 함수)
- `len()` 함수를 사용하면 문자열의 길이를 구할 수 있다.
```python
print("len('hello world'*1000)", len('hello world'*1000))
```
- 출력 결과:
```
len('hello world'*1000) 11000
```
- 즉, `'hello world'` 문자열을 1000번 반복한 결과는 총 **11000 글자**이다.


## 📘 문자열 치환 (`replace` 함수)
- 문자열의 일부를 다른 문자열로 바꾸고 싶을 때 `replace()` 함수를 사용한다.
```python
print("'hello world'.replace('world','universe')", 'hello world'.replace('world','universe'))
```
- 출력 결과:
```
'hello world'.replace('world','universe') hello universe
```
- `'world'`가 `'universe'`로 변경된 것을 확인할 수 있다.
