
# 📘 파이썬 데이터 타입의 개념 요약

## 🐍 1. 데이터 타입이란?

- 데이터 타입이란 **데이터들을 엄격하게 분류하는 기준**
- 예시:
  - `Number`: 숫자 데이터 → 예: `1`, `3.14`
  - `Text`: 문자 데이터 → 예: `'hello world'`
  - `Sequence`: 리스트, 튜플 등
  - `Boolean`: 참(True), 거짓(False)

✅ 데이터 타입은 **데이터의 성격**을 정의함

---

## 🐍 2. 왜 데이터 타입을 나누는가?

- **이유**: 데이터 타입에 따라 **연산 방식이 다르기 때문**

| 데이터 타입 | 예시 값         | 연산 방법       |
|-------------|------------------|-----------------|
| Number      | `1`, `3.14`      | `+`, `-`, `*`, `/` 등 수학 연산 |
| Text        | `'hello'`        | `len()` 등 문자열 연산 |

✅ 같은 기호라도 타입에 따라 **의미와 동작이 다름**

---

## 🐍 3. 핵심 개념

- **데이터 타입에 따라** 어떤 연산을 할 수 있는지가 달라짐
- 파이썬을 다루기 위해선:
  - 어떤 **데이터 타입**들이 있는지
  - 각 타입별로 어떤 **연산 방식**이 있는지

를 이해하는 것이 중요함

✅ 데이터 타입은 파이썬의 기본기를 구성하는 핵심 개념!
