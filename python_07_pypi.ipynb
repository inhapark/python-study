# 📘 pandas로 CSV 파일 읽기 + pip 설치 방법

🐍 **0. pandas란?**  
- **pandas**는 **표 형식(테이블)**의 데이터를 다루기 위한 파이썬 라이브러리  
✅ 엑셀처럼 **행(row)**과 **열(column)**을 기준으로 데이터를 분석/처리할 수 있음  

---

🐍 **1. pandas로 파일 읽기**
```python
import pandas
house = pandas.read_csv('house.csv')
print(house)
```
✅ CSV 파일을 불러와서 테이블 형태로 출력

---

🐍 **2. 데이터 미리보기 및 요약**
```python
print(house.head())       # 앞부분 데이터 출력
print(house.head(2))      # 앞의 2개 데이터 출력
print(house.describe())   # 데이터의 통계 요약 정보
```
✅ 복잡한 CSV 파일을 구조화하여 보기 좋게 정리해 줌

---

🐍 **3. pip와 PyPI 개념**
- **PyPI**: Python 패키지 저장소 (공식 웹사이트: [pypi.org](https://pypi.org))
- **pip**: PyPI에 등록된 패키지를 설치해주는 명령어 도구

✅ pip를 사용하면 원하는 라이브러리를 쉽게 설치 가능

```bash
pip install pandas
# 또는
python3 -m pip install pandas
```

---

👉 핵심 정리표

| 항목 | 설명 |
|------|------|
| `pandas` | 표 형식 데이터를 다루는 라이브러리 |
| `pandas.read_csv()` | CSV 파일 읽기 |
| `head()` | 앞부분 데이터 확인 |
| `describe()` | 데이터 통계 요약 |
| `pip install` | 패키지 설치 명령 |
| `PyPI` | 파이썬 공식 패키지 저장소 |
