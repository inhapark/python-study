# 📘 변수 활용법과 `print()` 출력 방식

🐍 **1. 기본 코드 예시**
```python
# 애완동물을 소개해 주세요~
animal = "강아지"
name = "연탄이"
age = 4
hobby = "산책"
is_adult = age >= 3

print("우리집 " + animal + "의 이름은 " + name + "예요")
print(name + "는 " + str(age) + "살이며, " + hobby + "을 아주 좋아해요")
print(name + "는 어른일까요? " + str(is_adult))
```

📌 **결과**
```
우리집 강아지의 이름은 연탄이에요  
연탄이는 4살이며, 산책을 아주 좋아해요  
연탄이는 어른일까요? True
```

✅ `age`가 3 이상이므로 `is_adult = True`

---

🐍 **2. 문자열, 숫자, 불리언 함께 출력하기**

1️⃣ **문자열과 변수 결합 시 `+` 연산자 사용**
```python
print("우리집 " + animal + "의 이름은 " + name + "예요")
```
✅ 문자열을 이어 붙이려면 `+`를 사용  
❗ 단, 문자열끼리만 결합 가능  

---

2️⃣ **정수형(int)·불리언(bool)은 `str()`로 변환해야 함**
```python
print(name + "는 " + str(age) + "살이며, " + hobby + "을 아주 좋아해요")
print(name + "는 어른일까요? " + str(is_adult))
```
✅ `str()`은 숫자나 불리언을 **문자열로 변환**하는 함수  
✅ `str(4)` → `"4"`, `str(True)` → `"True"`  
💡 쉽게 말해 **“숫자를 글자로 바꾸는 함수”**

---

🐍 **3. 변수의 재할당**
```python
animal = "강아지"
print(animal)

animal = "고양이"
print(animal)
```

📌 결과:
```
강아지
고양이
```
✅ 같은 변수 이름을 다시 사용하면 **새로운 값으로 덮어쓰기**됨  
✅ 즉, **아래쪽 코드의 값이 최종 채택되어 출력됨**

---

🐍 **4. `+` 대신 `,`(콤마)로 출력하기**
```python
print(name, "는", age, "살이며,", hobby, "을 아주 좋아해요")
```

📌 결과:
```
연탄이는 4 살이며, 산책 을 아주 좋아해요
```

✅ 콤마(`,`)는 각 요소 사이에 **자동으로 공백을 추가**  
✅ `str()`을 사용하지 않아도 정수형, 불리언형 등 **모든 자료형 출력 가능**  
✅ 단, 공백이 자동 삽입되므로 결과를 깔끔히 제어하기 어려울 수 있음  

---

🐍 **5. 핵심 정리**

| 구분 | 설명 | 예시 |
|------|------|------|
| 문자열 결합 | `+` 사용, `str()` 필요 | `"이름은 " + name + "입니다"` |
| 숫자·불리언 출력 | 반드시 `str()`로 변환 | `"나이는 " + str(age)` |
| 변수 재할당 | 같은 이름의 변수는 마지막 값으로 덮어씀 | `animal = "고양이"` |
| 콤마(,) 사용 | 자료형 자동 변환, 공백 포함 | `print(name, age, hobby)` |

---

✅ **정리**
- `+`는 문자열끼리만 연결할 수 있고, 정수나 불리언은 `str()`로 감싸야 함  
- 같은 변수 이름을 다시 지정하면 **가장 마지막에 입력한 값이 적용됨**  
- `,`(콤마)는 자동으로 자료형 변환을 처리하므로 `str()`이 필요 없지만, **공백이 자동으로 들어감**
