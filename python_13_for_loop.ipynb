# 📘 반복문 `for`와 리스트

🐍 **1. 반복문과 리스트의 관계**
- **리스트**: 여러 데이터를 묶어서 저장하는 자료형  
- **반복문**: 리스트의 원소를 **하나씩 꺼내서 처리**하는 구조  
✅ 반복문과 리스트는 서로 **상호보완적 관계**

---

🐍 **2. `for` 반복문의 기본 구조**
```python
for 변수 in 리스트:
    실행문
```
✅ 리스트의 각 원소를 변수에 하나씩 담아 실행문을 반복 실행함

---

👉 `for` 문 구성 요소 정리표

| 항목           | 설명                 | 예시                          |
| -------------- | ------------------ | ----------------------------- |
| `for`          | 반복을 시작하는 **키워드**   | `for name in names:`         |
| `in`           | 리스트 등에서 값을 꺼낸다는 의미 | `for x in [1, 2, 3]:`        |
| `:`            | 들여쓰기 블록 시작         | `for x in data:` 다음 줄에 들여쓰기 |
| `for ~ in ~`   | 반드시 짝으로 사용         | `for item in list:`          |

---

🐍 **3. 기본 예제**
```python
names = ['A', 'B', 'C']
for name in names:
    print(name)
```

📌 결과 출력:
```
A  
B  
C
```

✅ 리스트 `names`의 요소를 하나씩 꺼내 `name`에 담고 출력

---

🐍 **4. 심화 예제**
```python
names = ['egoing', 'basta', 'blackdew', 'leezche']
for name in names:
    print('Hello, ' + name + ' . Bye, ' + name + '.')
```

📌 결과 출력:
```
Hello, egoing . Bye, egoing.  
Hello, basta . Bye, basta.  
Hello, blackdew . Bye, blackdew.  
Hello, leezche . Bye, leezche.
```

✅ `name`이라는 변수에 각 이름이 담기고, 문자열 조합으로 인사 메시지 출력

---

👉 핵심 정리표

| 개념 | 설명 |
|------|------|
| 리스트 | 여러 데이터를 묶는 자료형 (`['A', 'B', 'C']`) |
| `for 변수 in 리스트:` | 리스트 요소를 반복적으로 처리 |
| 변수 | 리스트에서 하나씩 꺼낸 값이 담기는 자리 |
| 반복문 목적 | 데이터를 순차적으로 처리하기 위함 |
